```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
# Загружаем данные пользователей
users = pd.read_csv('info.csv')
```

```python
users.columns = users.columns.str.strip()
```

```python
# Пример данных нового пользователя (new_user) - замените на реальные данные
new_user = pd.DataFrame({
    "gender": [1],  # Например, 'F' = 1
    "1": [1],
    "2": [45],
    "3": [90],
    "4": [15],
    "5": [8],
    "6": [13],
    "7": [63],
    "8": [57],
    "9": [96],
    "10": [42],
    "11": [87],
    "12": [0.93]
})
```

```python
# Выбираем случайные
sample_users = users
```

```python
# Посмотрим на первые несколько строк каждого файла для анализа структуры данных
sample_users.head(), users.columns
```

```python
# Создаем датафрейм для корреляции, с признаками нового пользователя
new_user_features = new_user.drop(columns=['gender'], axis=1)  # Убираем столбец пола
sample_users_features = sample_users.drop(columns=['id','gender'], axis=1)  # Убираем столбец пола для расчета
```

```python
# Посмотрим на первые несколько строк каждого файла для анализа структуры данных
new_user_features.head()
```

```python
# Посмотрим на первые несколько строк каждого файла для анализа структуры данных
sample_users_features.head()
```

```python
# Инициализируем список для хранения значений корреляций
correlations = []
```

```python
# Рассчитываем корреляцию Пирсона между new_user и каждым пользователем из выборки
for index, user_row in sample_users_features.iterrows():
    corr = new_user_features.iloc[0].corr(user_row)  # Корреляция Пирсона между строками
    correlations.append((index, corr))
```

```python
correlations
```

```python
# Создаём DataFrame для корреляций
correlation_df = pd.DataFrame(correlations, columns=['user_index', 'pearson_correlation'])
```

```python
correlation_df.head()
```

```python
# Сортируем по корреляции по убыванию
correlation_df = correlation_df.sort_values(by='pearson_correlation', ascending=False)
```

```python
# Выводим топ-10 самых похожих пользователей
top_similar_users = correlation_df.head(10)
```

```python
top_similar_users
```

```python
# Загружаем таблицу рейтингов
ratings = pd.read_csv("rec-libimseti-dir.edges", delim_whitespace=True, skiprows=1, names=["id_from", "id_to", "rating"])
```

```python
# Новый датафрейм для хранения похожих пользователей
similar_user_ratings = []
```

```python
# Для каждого из топ-5 похожих пользователей ищем их id_from в ratings
for user_idx in top_similar_users['user_index']:
    user_id = users.iloc[user_idx]['id']  # Получаем id пользователя из исходной таблицы
    similar_ratings = ratings[ratings['id_from'] == user_id]  # Находим строки, где id_from совпадает с id пользователя
    # Фильтруем только те строки, где рейтинг >= 6
    similar_ratings_filtered = similar_ratings[similar_ratings['rating'] >= 6]
    # Добавляем id_to и rating в список
    similar_user_ratings.append(similar_ratings_filtered[['id_to', 'rating']])
```

```python
# Создаем финальный датафрейм из всех добавленных данных
final_similar_user_ratings = pd.concat(similar_user_ratings, ignore_index=True)
```

```python
# Сортируем финальный датафрейм по rating по убыванию и берем только 10 первых
final_similar_user_ratings_sorted = final_similar_user_ratings.sort_values(by='rating', ascending=False).head(10)
```
